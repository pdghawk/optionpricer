# Comparing Monte Carlo, and PDE methods for pricing call options on a single underlying



```python
import sys
sys.path.append('..')


from optionpricer import payoff
from optionpricer import option
from optionpricer import bspde
from optionpricer import analytics
from optionpricer import parameter as prmtr
from optionpricer import path
from optionpricer import generator
from optionpricer import montecarlo

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# choose the expiry and strike for the option
expiry = 1.0/12.0 # time until expiry (annualized)
strike = 50.0     # the strike price of the option

r0   = 0.02 # interest rate (as decimal, not percentage) 
sig0 = 0.12 # volatility (annualized)
# create SimpleParam's using prmtr - they have methods for getting mean etc
r_param   = prmtr.SimpleParam(r0) 
vol_param = prmtr.SimpleParam(sig0)

# spots at which to calculate option price
spots = np.linspace(40.0,70.0,30)
```

```python
# create payoff for call, and option using that payoff and specified expiry

payoff_ = payoff.CallPayOff(strike)
option_ = option.VanillaOption(payoff_,expiry)
```

```python
# for the monte carlo generator, specify a statistical generator for the underlying's log returns

# here we choose a normal distribution, as assumed in Black-Scholes world
gen_norm = generator.Normal()
# for variance reduction of the model, create an antithetic generator built onto the base generator
gen_norm_antith = generator.Antithetic(gen_norm)
```

```python
# create a montecarlo pricing object
mc_pricer = montecarlo.SAMonteCarlo(option_,gen_norm_antith)
# initialize array of option prices
mc_prices = np.zeros_like(spots)
# loop through the spots to sokve for, and solve the pricer
for s_idx,s in enumerate(spots):
    # note the optional tolerance requirement on the price
    mc_prices[s_idx]  = mc_pricer.solve_price(s,r_param,vol_param,eps_tol=0.0001)
```

```python
# for comparison, also calculate the prices via PDE methods

# first, create a boundary condition object for the solver
bcs = bspde.BoundaryConditions1D()
# we should really here apply specific boundary conditions, such as Dirichlet or Neumann
# conditions at the edges of the grid. Here though, in this simple case they can be neglected
# at the cost of some accuracy.

# create the pricing object
bspricer = bspde.BlackScholesSingleAssetPricer(option_,r0,sig0,bcs)
# solve for the spots of interest
pde_prices   = bspricer.solve(spots)
```

```python
# also for comparison - get the analytic (exact) option prices
bs_prices = analytics.black_scholes_call_price(spots,strike,expiry,r0,sig0)
```

```python
# set up the plotting paramters for nice figures
sns.set()
mpl.rcParams['lines.linewidth']  = 2.0
mpl.rcParams['font.weight']      = 'bold'
mpl.rcParams['axes.labelweight'] = 'bold'
mpl.rcParams['font.size']        = 12
mpl.rcParams['legend.frameon']   = False
```

```python
# plot the prices of the option for different spots
# via the different pricing methods
plt.fill_between(spots,bs_prices,facecolor='b',alpha=0.4,label='Analytic')
plt.plot(spots,pde_prices,'k',label='PDE')
plt.plot(spots,mc_prices,'r-.',label='Monte Carlo')
plt.plot([strike, strike],[0,1.2*np.amax(bs_prices)],'b',label='strike')
plt.legend(frameon=False)
plt.xlabel('spot')
plt.ylabel('option price')
plt.title(str(option_))
#
plt.xlim([spots[0], spots[-1]])
plt.ylim([0,1.2*np.amax(bs_prices)])

plt.show()
```

```python
# we can also compare the errors on the pricing methods
# by comparing the solutions with the anaytic price.

pde_error = np.abs(pde_prices-bs_prices)/bs_prices
mc_error  = np.abs(mc_prices-bs_prices)/bs_prices
plt.semilogy(spots,pde_error,'k',label='PDE')
plt.semilogy(spots,mc_error,'r-.',label='Monte')
plt.plot([strike, strike],[1e-9,1],'b',label='strike')
plt.legend(frameon=False)
plt.xlabel('spot')
plt.ylabel('|price - analytic|/analytic')
plt.title(str(option_))

plt.xlim([spots[0], spots[-1]])
plt.ylim([1e-9, 1])

plt.show()
```
